## EDA of Customers

```python
# Import package
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Fix view Chinese problem 
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['Microsoft YaHei']    # 指定預設字型:解決plot不能顯示中文問題
mpl.rcParams['axes.unicode_minus'] = False           # 解決儲存影象是負號'-'顯示為方塊的問題
```

```python
# print current working directory
print(os.getcwd())
```

```python
#Import data
PChome = pd.read_csv('Full_PChome_utf.csv', encoding = 'utf-8')
PChome = PChome.drop(['idx','postal_cd', 'date_cd', 'goods','Zip_Code'], axis = 1)
PChome['date_create'] = pd.to_datetime(PChome['date_create'])
PChome['year'] = PChome['date_create'].map(lambda x: x.year)
PChome['month'] = PChome['date_create'].map(lambda x: x.month)
PChome['day'] = PChome['date_create'].map(lambda x: x.day)
PChome.head()
```

```python
# Plot by month
value = PChome.groupby(PChome['month'])['P_Q'].sum()
ax = PChome.groupby(PChome['month'])['P_Q'].sum().to_frame().plot(kind = 'bar', color = 'lightblue')
ax.set_yticklabels(np.arange(0, max(value)+10000000, 10000000))
```

```python
# Plot by date
value = PChome.groupby(PChome['day'])['P_Q'].sum()
ax = PChome.groupby(PChome['day'])['P_Q'].sum().to_frame().plot(kind = 'bar', color = 'lightskyblue', figsize = (10,5))
ax.set_yticklabels(np.arange(0, max(value)+1000000, 1000000))
```

```python
max(PChome.groupby(PChome['department'])['P_Q'].sum())
```

```python
# Plot by department
value = PChome.groupby(PChome['department'])['P_Q'].sum()
ax = PChome.groupby(PChome['department'])['P_Q'].sum().sort_values(ascending=False)\
.to_frame().plot(kind = 'bar', color = 'seagreen', figsize = (10,5))
ax.set_title('Plot by department')
# ax.set_yticklabels(np.arange(0, 100000000, 1000000))
```

```python
# Plot by Region
ax = PChome.groupby(PChome['Region'])['P_Q'].sum().sort_values(ascending=False)\
.to_frame().plot(kind = 'bar', color = 'palegreen', figsize = (8,5))
ax.set_title('Plot by Region')
```
